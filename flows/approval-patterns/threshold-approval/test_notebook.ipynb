Auto-approve vs escalate based on amount thresholds.

## Test 1: Auto-Approve (<$50)

```python
amount = 30.00

if amount < 50:
    decision = "AUTO-APPROVED"
    approver = "System"
elif amount <= 500:
    decision = "ESCALATED"
    approver = "Manager"
else:
    decision = "ESCALATED"
    approver = "Director"

result = [{'Amount': f'${amount}', 'Decision': decision, 'Approver': approver}]
pd.DataFrame(result)
```

## Test 2: Manager Approval ($50-$500)

```python
amount = 250.00

if amount < 50:
    decision = "AUTO-APPROVED"
    approver = "System"
elif amount <= 500:
    decision = "ESCALATED"
    approver = "Manager"
else:
    decision = "ESCALATED"
    approver = "Director"

result = [{'Amount': f'${amount}', 'Decision': decision, 'Approver': approver}]
pd.DataFrame(result)
```

## Test 3: Director Approval (>$500)

```python
amount = 1500.00

if amount < 50:
    decision = "AUTO-APPROVED"
    approver = "System"
elif amount <= 500:
    decision = "ESCALATED"
    approver = "Manager"
else:
    decision = "ESCALATED"
    approver = "Director"

result = [{'Amount': f'${amount}', 'Decision': decision, 'Approver': approver}]
pd.DataFrame(result)
```
